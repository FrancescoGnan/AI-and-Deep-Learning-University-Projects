**De Jong Function 1 (Sphere Function)**

The de Jong function number 1, often referred to as the Sphere function, is a benchmark problem in optimization. The function is defined as:

$$ f(\mathbf{x}) = \sum_{i=1}^{n} x_i^2 $$

Where:
- $ f(\mathbf{x}) $ is the function value.
- $ x_i $ represents each dimension of the input vector $ \mathbf{x} $.
- $ n $ is the number of dimensions of the input vector $ \mathbf{x} $.

The function's global minimum is 0, which occurs when $ x_i = 0 $ for all $ i $. The Sphere function is convex and unimodal, making it relatively simple compared to other benchmark functions.

##De Jong Function 2 (Rosenbrock's Valley)

The de Jong function number 2, commonly known as the Rosenbrock's Valley or the Rosenbrock function, is another benchmark problem in optimization. The function is defined as:

$$ f(\mathbf{x}) = \sum_{i=1}^{n-1} \left[ 100(x_{i+1} - x_i^2)^2 + (1 - x_i)^2 \right] $$

Where:
-  $f(\mathbf{x})$ is the function value.
-  $x_i$ represents each dimension of the input vector $\mathbf{x}$.
- $n$ is the number of dimensions of the input vector $\mathbf{x}$.

The function has a global minimum of 0, which occurs at $x_i = 1$  for all $i$. The Rosenbrock function is non-convex and has a narrow, curved valley. The bottom of this valley is flat, and it is easy to find the valley but challenging to converge to the global minimum.

**De Jong Function 3 (Step Function)**

The de Jong function number 3, commonly referred to as the Step function, is another benchmark problem in optimization. The function is defined as:

$$ f(\mathbf{x}) = \sum_{i=1}^{n} \lfloor x_i \rfloor $$

Where:
- $ f(\mathbf{x}) $ is the function value.
- $ x_i $ represents each dimension of the input vector $ \mathbf{x} $.
- $ n $ is the number of dimensions of the input vector $ \mathbf{x} $.
- $ \lfloor \cdot \rfloor $ denotes the floor function, which rounds its argument down to the nearest whole number.

The function's global minimum is when the $\mathbf{x}$ is equal to (-5.12, -5.12, -5.12, ...) and $f(\mathbf{x}) = 0$. The Step function is non-continuous and non-differentiable, making it a challenging problem for some optimization algorithms.

> # Gray Code (Gray Encoding)
>
> Gray code is a binary numeral system where two successive numbers differ in only one bit. This makes it particularly useful in scenarios where it's beneficial to avoid sudden changes in multiple bits, such as analog-to-digital converters.
>
> ## Properties of Gray Code:
>
> 1. **Uniqueness:** Every binary number has a unique Gray code equivalent.
> 2. **Single-bit change:** Only one bit changes at a time when moving from one value to the next.
>
> ## How to Convert Binary to Gray Code:
>
> To convert a binary number to Gray code:
>
> 1. The MSB (most significant bit) of the Gray code is the same as the original number.
> 2. The subsequent bits of the Gray code can be obtained by XOR-ing the current bit with the next significant bit.
>
> For example, to convert the binary number `1101` to Gray code:
>
> 1. The MSB is `1`.
> 2. XOR the first and second bits: `1 ⊕ 1 = 0`.
> 3. XOR the second and third bits: `1 ⊕ 0 = 1`.
> 4. XOR the third and fourth bits: `0 ⊕ 1 = 1`.
>
> Thus, the Gray code representation is `1011`.
>
> ## How to Convert Gray Code back to Binary:
>
> 1. The MSB of the binary code is the same as the Gray code.
> 2. To get the subsequent binary bits, XOR the current Gray code bit with the previous binary bit.
>
> For example, to convert the Gray code `1011` back to binary:
>
> 1. The MSB is `1`.
> 2. XOR the MSB with the next Gray code bit: `1 ⊕ 0 = 1`.
> 3. XOR the previous result with the next Gray code bit: `1 ⊕ 1 = 0`.
> 4. XOR the previous result with the next Gray code bit: `0 ⊕ 1 = 1`.
>
> Thus, the binary representation is `1101`.

```
# It's a class that implements a genetic algorithm for the De Jong function
class GeneticAlgorithmForDJ:

  def __init__(self, dj_instance, selection_rule, elitism=False, population_array=None, population_size=20,
               mutation_rate=0.01, iterations=600):
    """
    This function initializes the genetic algorithm with the given parameters.

    :param dj_instance:      This is the instance of the De Jong model that you want to optimise.
    :param selection_rule:   The selection rule to be used. Can be either 'rank' or 'roulette'.
    :param elitism:          True or False. If True, the best individual in the population will be carried over
                             to the next generation. Defaults to False (optional), in this case only the population
                             is replaced by the new generation.
    :param population_array: Optional parameter, can contain an initial array with the population. It is a list of
                             lists, where each inner list correspond to an individual of the population.
    :param population_size:  The number of individuals in the population, defaults to 20 (optional).
    :param mutation_rate:    The probability of a gene mutating, defaults 0.01 (optional).
    :param iterations:       The number of iterations to run the algorithm for, defaults to 600 (optional).
    """

    self.dj_instance = dj_instance
    self.dimension = self.dj_instance.dimension
    self.dimension_bits = self.dj_instance.num_bits
    self.selection_rule = selection_rule
    self.population_size = population_size
    self.population_array = population_array
    self.mutation_rate = mutation_rate
    self.iterations = iterations
    self.elitism = elitism

    # Encoding functions
    self.encoding_function = np.vectorize(self.dj_instance.gray_encode)
    self.decoding_function = np.vectorize(self.dj_instance.gray_decode)

    # Statistics variables
    self.best_solution = None
    self.best_eval = None
    self.best_generation = None
    self.best_evals = None

    self.gen_mean_eval = None
    self.gen_error_eval = None

    # Additional variables

    self.fitness = None
    self.evaluations = None
    self.pop = None
    self.children = None

    self.nextpop = None
    self.nextfit = None
    self.nexteval = None

  def solve(self):
    """
    This function is the main of the class `GeneticAlgorithm`. It performs the main loop of the algorithm,
    and it calls the other functions of the class.
    """
    # Initialise population
    # ...
    self.pop = self.initialise_population()

    # Evaluate population
    # ...
    self.fitness, self.evaluations = self.compute_fitness(self.pop)
    #print(self.fitness)
    #print(self.evaluations)

    self.best_eval = min(self.evaluations)
    self.best_evals = np.array([min(self.evaluations)])

    self.best_solution = self.pop[np.argmin(self.evaluations)]
    self.best_generation = 0

    self.gen_mean_eval = np.array([np.mean(self.evaluations)])
    self.gen_error_eval = np.array([np.std(self.evaluations)])


    for iteration in range(self.iterations):

      if self.selection_rule == 'rank':
            prob = self.rank_selection()
      elif self.selection_rule == 'roulette':
            prob = self.roulette_selection()

  # Select parents
  # ...
      parents = self.selection( prob )

  # Generates new individuals using reproduction rules
  # ...
      self.crossover( parents )
      #print(len(parents))
      self.mutation()

      #print(len(self.pop))
      #print(len(self.children))
      #print(len(parents))

      self.reconstruction()

  # Form a new population
  # ...

  # Evaluate population
  # ordinare in senso opposto!!!!!!!
  # ...

      #print(self.nextpop)
      self.update_statistics(iteration)

  def initialise_population(self):
    """
    This function generates a random population of shape `population_size` and `dimension` with values between
    `range[0]` and `range[1]`.
    :return: The population encoded using gray code.
    """
    N_pop = self.population_size
    dim_pop = self.dimension
    r0 = self.dj_instance.range[0]
    r1 = self.dj_instance.range[1]

    pop = np.zeros((N_pop, dim_pop))

    for i in range(N_pop):
      for j in range(dim_pop):
        pop[i][j] = rnd.uniform(r0, r1)

    pop = pop.reshape(self.population_size,self.dimension)
    return self.encoding_function(pop)

  def compute_fitness(self, pop):
    """
    :return:
    """
    eval = self.dj_instance.evaluate(pop, gray_=True)

    # We normalize the evaluations and we add a small value epsilon
    # to avoid possible divisions by zero
    fit = max(eval) - eval + np.finfo(float).eps
    fit = fit / max(fit)

    return fit, eval

  def rank_selection(self):
    """
    This function compute the individual probability of selection using rank selection rule.
    :return: The probability of each individual being selected for the next generation.
    """
    N_pop = self.population_size

    ranked_pop = np.argsort(self.fitness) #returns the indeces of the sorted-by-rank array

    sum_j = 0                        #denominator of P(i)
    for i in range(1, N_pop+1):
      sum_j += 1

    P = [] # P(i) is the selection probablility of the individual with rank i
    for i in range(N_pop):
      rank_i = i / sum_j
      P.append( rank_i )

    P_worst = 2 / ( N_pop * (N_pop + 1) )
    P_best = 2 / ( N_pop + 1 )

    #selected_individuals = []
    #for i in range(N_pop):
    #    if np.random.rand() < P[i]:
    #        selected_individuals.append(self.population_array[ranked_pop[i]])

    #return selected_individuals
    return P

  def roulette_selection(self):
    """
    This function compute the individual probability of selection using roulette selection rule.
    :return: The probability of each individual being selected for the next generation.
    """

    fitness_tot = sum( self.fitness )
    P = [ fit / fitness_tot for fit in self.fitness ]

    return P

  def selection(self, P):
    """
    This function is used to select the parents for the next generation according to a selection rule.
    """
    # Computing the probability of each individual to be selected as parent.
    # ...

    # Selecting the parents for the next generation using probabilities.
    # ...
    N = self.population_size
    parents = []

    for _ in range( N ):
        n = rnd.random()
        acc_P = 0
        for i in range( N ):
            acc_P += P[i]
            if n <= acc_P:
                parents.append(i)
                break

    return parents

  def crossover(self, parents):
    """
    This function perform a single-point crossover on each pair of parents, selected by pairing the first half of
    the parents with the second half of them.
    """
    self.children = []
    for i in range(0, len(parents), 2):  # step size of 2
        child_1 = []
        child_2 = []

        parent1 = self.pop[parents[i]]
        parent2 = self.pop[parents[i + 1]]

        crossover_point = rnd.randint(0, len(parent1))

        child_1.extend(np.concatenate((parent1[:crossover_point], parent2[crossover_point:])))
        child_2.extend(np.concatenate((parent2[:crossover_point], parent1[crossover_point:])))

        self.children.extend([child_1, child_2])

    self.children = np.array(self.children.copy())


  def mutation(self):
    """
    This function, decide if mutate or not each bit of each offspring in each child, by flip it, with a mutation
     rate, otherwise keep it the same.
    """
    self.nextpop = [child.copy() for child in self.children]

    for child in self.nextpop:
        for i in range(len(child)):
            mutation = ''
            for value in child[i]:
                if np.random.uniform() < self.mutation_rate:
                    mutation += str(1 - int(value))
                else:
                    mutation += value
            child[i] = mutation

    self.nextfit, self.nexteval = self.compute_fitness(self.nextpop)

  def elitism_selection(self):
    """
    Generating a new population as combination of the old one and the new one.
    fit_parents = self.dj_instance.evaluate(parents, gray_=True)
    fit_children = self.dj_instance.evaluate(children, gray_=True)
    best_index_parents = sorted(fit_parents, reverse=True)
    best_index_children = sorted(fit_children, reverse=True)
    best_index_parents = best_index_parents[:n]
    best_index_children = best_index_children[:n]
    best_n_parents = []
    best_n_children = []
    for i in range( len(n) ):
      best_n_parents.append( parents[best_index_parents[i]] )
      best_n_children.append( children[best_index_children[i]] )

    final_candidates = best_n_parents + best_n_children
    return final_candidates
    """
    # Concateno pop e nextpop, le ordino e prendo la migliore meta'
    double_pop = np.concatenate([ self.pop, self.nextpop ])
    double_fit, double_eval = self.compute_fitness(double_pop)
    return double_fit, double_pop

  def reconstruction(self):
    """
    If elitism is enabled, this function replace the population with the 20 best individuals.
    """

    #elite_candidates = self.elitism_selection( parents, children, n )
    #self.population_array = new_pop
    #self.compute_fitness()
    #sorted_index = sorted(self.fitness, reverse=True)

    #new_pop[ -n: ] = elite_candidates

    if self.elitism == False:
      self.pop = self.nextpop
    else:
      double_fit, double_pop = self.elitism_selection()
      fit_and_indices = []
      for i in range(len(double_fit)):
          fit_and_indices.append((double_fit[i], i))
      fit_and_indices.sort(reverse=True)
      selected_indices = [fit_index[1] for fit_index in fit_and_indices[:self.population_size]]
      self.pop = [double_pop[index] for index in selected_indices]

  def update_statistics(self,iteration):
    """
    This function can be used to compute statistics on the objective value.
    """
    self.fitness, self.evaluations = self.compute_fitness(self.pop)

    min_eval_index = np.argmin(self.evaluations)
    min_eval = self.evaluations[min_eval_index]
    best_candidate = self.pop[min_eval_index]

    self.gen_mean_eval = np.append(self.gen_mean_eval, np.mean(self.evaluations) )
    self.gen_error_eval = np.append(self.gen_error_eval, np.std(self.evaluations) )
    self.best_evals = np.append(self.best_evals, min( min_eval, self.best_eval ) )

    if min_eval < self.best_eval:
        self.best_generation = iteration
        self.best_eval = min_eval
        self.best_solution = best_candidate
```

In the following cell we define the parameters of our model as requested.
We are going to use a population size of 20 individuals and to evolve them through 600 generations. At each generation we select the best chromosomes according to their fitness using to different rules: rank and roulette selection.

As far as rank rule is concerned, it assigns probabilities for selecting individuals from a population based on their rank in terms of fitness. In particular, the fittest individuals are assigned the highest ranks.

The idea behind the roulette selection is quite similar. We imagine to spin a roulette wheel in order to choose an individual in the population: the higher the fitness, the larger the corresponding section on the roulette wheel, the higher the probability for the specific chromosome to be selected.

Moreover, we decide whether to use the elitism strategy or not. This means that at each generation we copy a certain number of best individuals between parents and children directly into the next generation of candidates. Despite this choice reduces the diversification, i.e. the ability of the algorithm to explore the space of solution, we studied during the theoretical lectures that this method improves the performances of the search.
Thus, we expect our code to achieve better results when elitism is taken into account.

Finally, at each iteration, once the parents are selected according to the probabilities extracted using rank/roulette rule, we perform the crossover operation to create children candidates that inherit genes from both their two parents. Each child has a certain probabiliy to be slightly mutated in order to improve diversification.

Looking at the table above we understand that as expected our code performs differently when we use different seeds.
In fact during the search our algorithm extract random numbers in order to decide for example whether to mutate an individual or which parents to select.
We could benefit from this fact by implementing the GA in parallel. Each core researches starting from a different random seed and after a certain number of iteration it shares its best individuals with the other cores.

---
In conclusion, we want to average the statistics on seed.

```
c_labels2 = ['Function', 'Method', 'Best mean of objective values']
small_data = []

for function in functions:
    fig, axes = plt.subplots(nrows=1, ncols=len(methods), figsize=(25, 8))
    fig.suptitle(f'{function.function_name}')

    method = 0
    for r_idx, sr in enumerate(selection_rules):
        for e in bool_elitism:
            mean_values = []
            std_values = []
            best_eval_values = []

            for seed in seeds:
                np.random.seed(seed)

                GA = GeneticAlgorithmForDJ(dj_instance=function, selection_rule=sr, elitism=e)
                GA.solve()

                mean_values.append(GA.gen_mean_eval)
                std_values.append(GA.gen_error_eval)
                best_eval_values.append(GA.best_eval)

            mean = np.mean(mean_values, axis=0)
            std = np.mean(std_values, axis=0)
            best_eval = np.mean(best_eval_values)

            ax = axes[method]
            ax.set_title(f'Method: {methods[method]}')
            x = np.arange(GA.iterations + 1)
            ax.plot(x, mean, color='green', label='Mean')
            ax.fill_between(x, mean - std, mean + std, color='lightblue', label='Standard deviation')

            if function.function_name == 'De Jong 2' or function.function_name == 'De Jong 5':
                ax.set_ylim(0, 1000)
            else:
                ax.set_ylim(0, 60)

            ax.set_xlabel('Generations')
            ax.set_ylabel('Evaluation function')
            ax.legend()

            small_data.append([function.function_name, methods[method], best_eval])

            method += 1

# plt.savefig('images/' + function.function_name + '.png', dpi=600)
plt.show()

small_df = pd.DataFrame(small_data, columns=c_labels2)
```

In conclusion, in order to evaluate the performance of a genetic algorithm, it is a good practice to run it starting from different random seeds. Then an estimation of its best solution is given by the mean of the best solutions obtained with each seed.
